# Άσκηση 6 

```python
import numpy as np
from matplotlib import pyplot as plt
```

# h =0.1

```python
#vima
h = 0.1
N = (2-0)/h
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,2,int(N))
t
```

```python
# analytiki lysi
y = 1000/ (9*np.exp(-10*t) +1)
y
```

```python
plt.plot(t,y, label ='analytiki lysi')
plt.legend() 
plt.show()
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=100
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h* 10*yeuler[i]*(1-yeuler[i]/1000)
yeuler
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.legend()
plt.show()
```

# Άσκηση 10 

```python
#theorw [0,1]
h=0.1
N = (1-0)/h
x = np.linspace(0,1,int(N))
x
```

```python
y = np.exp(x) -x -1 #analytiki lysi
y
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=0
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h*(x[i]+y[i])
yeuler
```

```python
#plot mazi
plt.plot(x,y, label ='analytiki')
plt.plot(x,yeuler, label ='euler')
plt.legend()
plt.show()
```

## Taylor

```python
ytaylor = np.zeros(int(N))
ytaylor[0]=0
for i in range(0,int(N-1)):
    ytaylor[i+1] = ytaylor[i] + h* (x[i]+y[i]) + (h**2/2)*(1+x[i]+y[i])
ytaylor
```

```python
#plot mazi
plt.plot(x,y, label ='analytiki')
plt.plot(x,yeuler, label ='euler')
plt.plot(x,ytaylor, label ='taylor')

plt.legend()
plt.show()
```

```python
dn_taylor = y - ytaylor
dn_taylor
```

```python
dn_euler = y - yeuler
dn_euler
```
