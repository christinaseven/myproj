# Άσκηση 2

```python
import numpy as np
import math 
from matplotlib import pyplot as plt
```

```python
#vima
h = 0.1
N = (2-0)/h
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,2,int(N))
t
```

```python
# analytiki lysi
y = 1000/ (9*np.exp(-10*t) +1)
y
```

```python
plt.plot(t,y, label ='analytiki lysi')
plt.legend() 
plt.show()
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=100
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h* 10*yeuler[i]*(1-yeuler[i]/1000)
yeuler
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.legend()
plt.show()
```

```python
#rk4 lysi
yrk = np.zeros(int(N))
yrk[0]=100
for i in range(0,int(N-1)):
    k1 = 10*yrk[i]*(1 - yrk[i]/1000)
    k2 = 10*(yrk[i]+h*(k1/2))*(1 - (yrk[i]+h*(k1/2))/1000)
    k3 = 10*(yrk[i]+h*(k2/2))*(1 - (yrk[i]+h*(k2/2))/1000)
    k4 = 10*(yrk[i]+h*k3)*(1 - (yrk[i]+h*k3)/1000)
    yrk[i+1] = yrk[i] + (1/6)*h*(k1+2*k2+2*k3+k4)
yrk
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,yrk, label ='rk4')

plt.legend()
plt.show()
```

# Άσκηση 5

## h =0.1 

```python
#vima
h = 0.1
N = (10-0)/h
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,10,int(N))
t
```

```python
t.shape # gia vima 0.1 douleuw me 100 shmeia (ara 100 sinartisiakoi upologismoi --> gia kathe ena shmeio)
```

```python
# analytiki lysi
y = np.exp(np.sin(t))
y
```

```python
plt.plot(t,y, label ='analytiki lysi')
plt.legend() 
plt.show()
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=1
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h* yeuler[i]* np.cos(t[i])
yeuler
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.legend()
plt.show()
```

```python
#rk4 lysi
yrk = np.zeros(int(N))
yrk[0]=1
for i in range(0,int(N-1)):
    k1 = yrk[i]*np.cos(t[i])
    k2 = (yrk[i]+h*(k1/2))*np.cos(t[i] + h/2)
    k3 = (yrk[i]+h*(k2/2))*np.cos(t[i] + h/2)
    k4 = (yrk[i]+h*k3)*np.cos(t[i] + h)

    yrk[i+1] = yrk[i] + (1/6)*h*(k1+2*k2+2*k3+k4)
yrk
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,yrk, label ='rk4')

plt.legend()
plt.show()
```

```python
#mesou lysi
ym = np.zeros(int(N))
ym[0]=1
for i in range(0,int(N-1)):
    ym[i+1] =(ym[i] + (h/2)*y[i]*np.cos(t[i]+(h/2)))*(1 / (1-(h/2)*np.cos(t[i]+(h/2))))
ym
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,yrk, label ='rk4')
plt.plot(t,yrk, label ='mesou')


plt.legend()
plt.show()
```

## Σφάλματα

```python
d_euler = np.abs(y - yeuler)
d_mesou = np.abs(y - ym)
d_rk = np.abs(y - yrk)
```

```python
print(f" The error of Euler in point 10 is: {d_euler[99]}")
print(f" The error of Mesou in point 10 is: {d_mesou[99]}")
print(f" The error of RK4 in point 10 is: {d_rk[99]}")
```

```python
plt.plot(t,d_euler, label ='euler')
plt.plot(t,d_mesou, label ='mesou')
plt.plot(t,d_rk, label ='rk4')
plt.title("Error comparison")

plt.legend()
plt.show()
```

# Άσκηση 11 

Θεωρώ ομοιόμορφη διαμέριση στο [0,1]. Επιλέγω βήμα 0.01.

```python
#vima
h = 0.01
N = (1-0)/h
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,1,int(N))
t
```

```python
# analytiki lysi
y = -(5/6)*np.power(t,3) -(1/6)*t +1
y
```

```python
plt.plot(t,y, label ='analytiki lysi')
plt.legend() 
plt.show()
```

```python
print(f" y(t0) = y0 is the value of first point : {y[0]}")
print(f" y(t1) = y1 is the value of the second point: {y[1]}")
```

```python
# peperasmenwn diaforwn lysi
yp = np.zeros(int(N))
yp[0]=1
yp[1] = y[1]
for i in range(0,int(N-2)):
    yp[i+2] = 2*yp[i+1] - yp[i] - 5* h*h* t[i+1]
yp
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yp, label ='peperasmenwn diaforwn')
plt.legend()
plt.show()
```

# Άσκηση 12

Θεωρώ ομοιόμορφη διαμέριση στο [0,1]. Επιλέγω βήμα 0.01.

```python
#vima
h = 0.01
N = (1-0)/h
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,1,int(N))
t
```

```python
# analytiki lysi
c1 = (-1/2) - (- np.exp(np.sqrt(2)) + np.exp(2* np.sqrt(2)))/ (2* (1 - np.exp(2*np.sqrt(2))))
c2 = (- np.exp(np.sqrt(2)) + np.exp(2* np.sqrt(2)))/ (2* (1 - np.exp(2*np.sqrt(2))))

y = c1 * np.exp(np.sqrt(2)*t) + c2 * np.exp(-np.sqrt(2)*t) +(1/2)
y
```

```python
plt.plot(t,y, label ='analytiki lysi')
plt.legend() 
plt.show()
```

```python
print(f" y(t0) = y0 is the value of first point : {y[0]}")
print(f" y(t1) = y1 is the value of the second point: {y[1]}")
```

```python
# peperasmenwn diaforwn lysi
yp = np.zeros(int(N))
yp[0]=0
yp[1] = y[1]
for i in range(0,int(N-2)):
    yp[i+2] = -h*h - yp[i] + (2*h*h +2)* yp[i+1]
yp
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yp, label ='peperasmenwn diaforwn')
plt.legend()
plt.show()
```
