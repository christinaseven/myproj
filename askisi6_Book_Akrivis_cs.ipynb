```python
h=0.1
#h=0.2
#h=0.25
#h=0.4
N= int(2/h)
import numpy as np;
t=np.linspace(0, 2, N+1)
y=np.zeros(N+1)
y[0]=100
#apo 0 mexri kai N
print("y0 = 100")
for n in range(0,N):
    y[n+1]= y[n]+ h * 10*y[n]*(1- (y[n]/1000))
    print("for n=", n , "euler solution is: y",n+1, "is:" , y[n+1])
```

Υπολογισμός προσεγγιστικών τιμών

```python
from matplotlib import pyplot as plt
plt.plot(y,t, label='approx')
#plt.xlabel("Value of t")
plt.ylabel("value of solution y")
plt.title("Approximate solution with euler's method")
plt.show()
```

```python
import math;
N= int(2/h)
ysol=np.zeros(N+1)
for i in range(0, N+1):
    ysol[i]=1000/(9*math.exp(-10*t[i])+ 1)
    print (ysol[i])
```

```python
plt.plot(ysol,t, label='analytical')
#plt.xlabel("Value of t")
plt.ylabel("value of solution y")
plt.title("Solution in analytical method")
plt.show()
```

```python
# Function to plot   
plt.plot(y,t, label ='eulers') 
plt.plot(ysol,t, label ='analytical') 
  
# Function add a legend   
plt.legend() 
# function to show the plot 
plt.show()
```
