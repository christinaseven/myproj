```python
import numpy as np
import math
import matplotlib.pyplot as plt 
```

```python
#Steps
#1. t in [0,2]
h=0.01
N = int((2-0)/h)
ti = np.linspace(0, 2, N)
ti
```

```python
#2. analytical solution
#y(a) = y(0) =100, [a,b] = [0,2]
def analytical(t):
    y = 1000/ (9* np.exp(-10*t) +1)
    return y
```

```python
y=analytical(ti)
y
```

```python
# ti -> x axis values
# y ->corresponding y axis values
 
# plotting the points
plt.plot(ti, y)
 
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
 
# giving a title to my graph
plt.title('analytical')
 
# function to show the plot
plt.show()
```

Σκοπός μας είναι να βρούμε την προσεγγιστική λύση με Euler και να τις συγκρίνουμε.

```python
# y_n+1 = y_n +10*h*y_n*(1-y_n/1000) h methodos
#y_0 = y(0) = y(a) =100
def euler(h,y0):
    y_euler = [] #list
    y_euler.append(y0)
    y_n = y0
    for i in range(0,N-1):
        y_n1 = y_n + 10 * h * y_n * (1 - y_n/1000)
        y_euler.append(y_n1)
        y_n = y_n1
    
    return np.array(y_euler)
```

```python
y_euler = euler(h,100)
y_euler.shape
y_euler
```

```python
#plotting
plt.plot(y_euler,ti, label ='eulers') 
plt.plot(y,ti, label ='analytical') 
# Function add a legend   
plt.legend() 
# function to show the plot 
plt.show()
```
