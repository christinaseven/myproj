```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

# h =0.1

```python
#vima
h = 0.1
N = (2-0)/h
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,2,int(N))
t
```

```python
# analytiki lysi
y = 1000/ (9*np.exp(-10*t) +1)
y
```

```python
plt.plot(t,y, label ='analytiki lysi')
plt.legend() 
plt.show()
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=100
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h* 10*yeuler[i]*(1-yeuler[i]/1000)
yeuler
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.legend()
plt.show()
```

# Άσκηση 10 

```python
#theorw [0,1]
h=0.1
N = (1-0)/h
x = np.linspace(0,1,int(N))
x
```

```python
y = np.exp(x) -x -1 #analytiki lysi
y
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=0
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h*(x[i]+y[i])
yeuler
```

```python
#plot mazi
plt.plot(x,y, label ='analytiki')
plt.plot(x,yeuler, label ='euler')
plt.legend()
plt.show()
```

## Taylor

```python
ytaylor = np.zeros(int(N))
ytaylor[0]=0
for i in range(0,int(N-1)):
    ytaylor[i+1] = ytaylor[i] + h* (x[i]+y[i]) + (h**2/2)*(1+x[i]+y[i])
ytaylor
```

```python
#plot mazi
plt.plot(x,y, label ='analytiki')
plt.plot(x,yeuler, label ='euler')
plt.plot(x,ytaylor, label ='taylor')

plt.legend()
plt.show()
```

```python
dn_taylor = y - ytaylor
dn_taylor
```

```python
dn_euler = y - yeuler
dn_euler
```

# Άσκηση 9 

```python
h=0.1
N = (2*math.pi -0)/h
t = np.linspace(0, 2*math.pi, int(N))
t
```

```python
#analytiki
y = np.cos(t)
y
```

```python
plt.plot(t, y, label = 'analytiki')
plt.legend()
plt.show()
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=1
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h*(1/2 - (y[i]**2)/2)
yeuler
```

```python
#plot mazi
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')

plt.legend()
plt.show()
```

```python
# mesou 
ymesou = np.zeros(int(N))
ymesou[0]=1
for i in range(0,int(N-1)):
    ymesou[i+1] = ymesou[i] + h*(1/2)*(1-(ymesou[i]/2 +(1-ymesou[i]**2)/4)**2)
ymesou
```

```python
#plot mazi
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,ymesou, label ='mesou')

plt.legend()
plt.show()
```

πλήρης αποτυχία. Είναι δύσκολη η προσέγγιση τριγωνομετρικών!

## Άσκηση 7

### euler

```python
#vima
h = 0.1
N = (2-0)/h #[0,2]
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,2,int(N))
t
```

```python
# analytiki lysi
y = np.exp(-t)
y
```

```python
plt.plot(t,y, label ='analytiki lysi')
plt.legend() 
plt.show()
```

```python
# euler lysi
yeuler = np.zeros(int(N))
yeuler[0]=1
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h* (-y[i])
yeuler
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.legend()
plt.show()
```

#### peplegmeni euler

```python
#vima
h = 0.1
N = (2-0)/h #[0,2]
```

```python
#omoiomorfi diamerisi
t = np.linspace(0,2,int(N))
t
```

```python
# peplegmeni euler 
ypeuler = np.zeros(int(N))
ypeuler[0]=1
for i in range(0,int(N-1)):
    ypeuler[i+1] = ypeuler[i]/(1+h) 
ypeuler
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,ypeuler, label ='peplegmeni')

plt.legend()
plt.show()
```

```python
# trapeziou 
trapeziou = np.zeros(int(N))
trapeziou[0]=1
for i in range(0,int(N-1)):
    trapeziou[i+1] = (trapeziou[i]-(h/2)*trapeziou[i])*(1/(1+(h/2)))
ypeuler
```

```python
plt.plot(t,y, label ='analytiki')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,ypeuler, label ='peplegmeni')
plt.plot(t,trapeziou, label ='trapeziou')


plt.legend()
plt.show()
```
